# Image Processing with PIL

```python
from PIL import Image
from IPython.display import display
```

```python
image = Image.open("./images/Healthy MRI Raw.png")
image # this line to displays the image in the notebook
```

### Cropping

```python
box = (100, 100, 400, 400)
cropped_image = image.crop(box)
cropped_image
```

### Resizing

```python
resized_image = image.resize((256,256))
resized_image
```

### Rotating

```python
rotated_image = image.rotate(45)  # Rotate by 45 degrees
rotated_image
```

### Enhancements

```python
from PIL import ImageEnhance

# Enhance the contrast
enhancer = ImageEnhance.Contrast(image)
contrast_enhanced_image = enhancer.enhance(2)  # Factor > 1 increases contrast
contrast_enhanced_image
```

### Filters

```python
from PIL import ImageFilter

blurred_image = image.filter(ImageFilter.BLUR)
embossed_image = image.filter(ImageFilter.EMBOSS)
edge_enhanced_image = image.filter(ImageFilter.FIND_EDGES)

display(blurred_image)
display(embossed_image)
display(edge_enhanced_image)
```

### Histograms

```python
histogram = image.histogram()
histogram
```

### Mirroring

```python
flipped_image_horizontal = image.transpose(Image.FLIP_LEFT_RIGHT)
flipped_image_vertical = image.transpose(Image.FLIP_TOP_BOTTOM)

display(flipped_image_horizontal)
display(flipped_image_vertical)
```

### Annotating

```python
from PIL import ImageDraw

draw = ImageDraw.Draw(image)
draw.text((10, 10), "Sample MRI", fill="white")

image
```

These are just a few examples of the many operations and transformations available in Pillow. Depending on the specifics of the MRI image and what you want to achieve, some operations may be more relevant than others. You can experiment with combinations of these operations and parameters to gain deeper insights from the brain scan or to create visual representations suited to your goals.
